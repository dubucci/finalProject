### Load Data

```python
df = pd.read_csv(os.path.join(job_dir + job_list[4]), encoding='cp949')
df
```

```python
years1=[]
years2=[]
for year in range(2016, 2021):
    y1 = '{}.7'.format(year)
    y2 = '{}.6'.format(year)
    years1.append(y1)
    years2.append(y2)
years2
```

```python
df_hire_rate = pd.concat([df.loc[1:, '시도별'], df.loc[1:, years1]], axis=1).reset_index().drop('index', axis=1)
df_fire_rate = pd.concat([df.loc[1:, '시도별'], df.loc[1:, years2]], axis=1).reset_index().drop('index', axis=1)
```

```python
# 결측치 제거 
df_hire_rate = df_hire_rate.replace('-', 0)
df_fire_rate = df_fire_rate.replace('-', 0)
```

```python
# 데이터 타입 변환 
df_hire_float = pd.concat([df_hire_rate.iloc[:, 0], df_hire_rate.iloc[:,1:].astype(float)], axis=1)
df_fire_float = pd.concat([df_fire_rate.iloc[:, 0], df_fire_rate.iloc[:,1:].astype(float)], axis=1)
```

```python
df_hire_float.replace('제주도', '제주특별자치도', inplace=True)
df_fire_float.replace('제주도', '제주특별자치도', inplace=True)
```
