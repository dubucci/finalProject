```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import MinMaxScaler
import os
```

```python
qualityfinal_dir = './final_data'
qualityfinal_list = os.listdir(qualityfinal_dir)
pd.DataFrame(qualityfinal_list)
```

```python
def inverse_min_max(df):
    df_min = df.min()
    df_max = df.max()

    return (df-df_max)/(df_min-df_max)
```

```python
df02 = pd.read_csv(os.path.join(qualityfinal_dir, qualityfinal_list[1]))
df02 = round(df02, 3)
df02.rename(columns = {'Unnamed: 0':'시도별'},inplace = True)
```

```python
df02
```

```python
df02_0 = df02.copy()
```

```python
for year in range(2000, 2021):
    df02_0[f'{year}'] = inverse_min_max(df02[f'{year}'])
```

```python
df02_0
```

```python
df02_0 = round(df02_0,3)
```

```python
df02_0 = pd.DataFrame(df02_0.stack(-1)).reset_index()
```

```python
df02_0
```

```python
df02_0 = df02_0.loc[df02_0['level_1']!='시도별', :].reset_index(drop=True)
```

```python
df02_0.rename(columns={'level_0':'CITY_CODE', 'level_1':'YEAR', 0:'DETAIL_INDEX'}, inplace=True)
```

```python
df02_0
```

```python
df02_0['DETAIL_ITEM_CODE'] = '비정규직 비율'
```

```python
df02_0
```

```python
df02_0.info()
```

```python
df02_0 = df02_0.astype({'DETAIL_INDEX':'float'})
```

```python
df02_0.to_csv('./final_index/df02.csv', index = False)
```
