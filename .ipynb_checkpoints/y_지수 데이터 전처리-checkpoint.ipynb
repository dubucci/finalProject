```python
def inverse_min_max(df):
    df_min = df.min()
    df_max = df.max()

    return (df-df_max)/(df_min-df_max)
```

```python
df_7_4 = df_7_2.copy()
```

```python
for year in range(2000, 2021):
    df_7_4[f'{year}'] = inverse_min_max(df_7_2[f'{year}'])
```
